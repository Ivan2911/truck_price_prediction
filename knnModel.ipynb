```python
data = pd.read_csv('truck_dataset.csv')
```

```python
#Count percentage of 0 price
count_zero = 0
for p in data['price']:
    if p ==0:
        count_zero +=1

zero_price = count_zero*100/len(data)
print('Zero price proportion is ' + str(zero_price) +'%') 
```
